# Blood Work Analysis

**Workflow**

``` python
Prompt (blood Report) -> LLM 1 (Extract Numbers) -> Extracted Numbers -> LLM 2 (Generate diet plan) -> diet plan
```

````python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
````

````python
with open("blood_work.txt", "r") as f:
    blood_report = f.read()

print(blood_report[:200])
````

````python
extraction_prompt = f"""
You are a medical expert. You are given a blood work report. Please extract the following information from the report:
1. Patient Name
2. Patient Age
3. Patient Gender
4. Date of the report
5. List of all blood test results with their corresponding values and reference ranges.

Extract ALL the test values and classify each one as HIGH, LOW or NORMAL 
based on the reference ranges provided in the report.

Format your response as:
- Test Name: value | Status: GIGH/LOW/NORMAL | Reference: range

Blood Report:
{blood_report}
"""

extraction_response = llm.invoke(extraction_prompt)
extracted_values = extraction_response.text

print("STAGE 1: Extracted Values")
print(extracted_values)
````

````python
diet_prompt = f"""
    You are a clinical nutritionist. You are given a blood work report and the extracted values from the report.
    Based on the extracted values, please provide a detailed diet plan for the patient.

    Write a short 2 lines explaining patient's current health status based on the blood work report and extracted values.
    
    The diet plan should include:
    1. Breakfast options
    2. Lunch options
    3. Dinner options
    4. Snacks options

    Blood Work Analysis:
    {extracted_values}
"""

diet_response = llm.invoke(diet_prompt)
diet_values = diet_response.text

print("=== STAGE 2: HEALTH SUMMARY & DIET PLAN ===")
print(diet_values)
````
