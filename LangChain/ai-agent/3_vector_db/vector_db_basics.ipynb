# Vector DB

```python
- VectorDB allows you to search through the meaning
- RDBMS allows you to search you to search through the exact values
```

````python
import chromadb
 
client = chromadb.Client()
````

````python
collection =  client.create_collection("news_articles")
````

````python
collection.add(
    ids=["1", "2", "3", "4", "5"],
    documents=[
        "Apple is looking at buying U.K. startup for $1 billion",
        "Tesla booked a minor profit of 1 billion $ in Q2",
        "Autonomous cars shift insurance liability toward manufacturers",
        "San Francisco considers banning sidewalk delivery robots",
        "SpaceX go NASA contract worth 10 billion $",
    ]
)
````

````python
# Fetch documents + embeddings for a given set of ids
results = collection.get(
    include=["documents", "embeddings", "metadatas"]
)

# Print Embeddings
for i, embedding in enumerate(results["embeddings"]):
    print(f"\n Document {i + 1}")
    print(f"Text: {results['documents'][i]}")
    print(f"Length of Embedding: {len(embedding)}\n")
    print(f"First 10 values: {embedding[:10]}\n")
````

````python
results = collection.query(
    query_texts=["This is query related to Elon Musk"],
    n_results=2
)

print(f"\nQuery Results: {results}") # look at the distance [[1.5531251430511475, 1.6894248723983765]] they are closely related
````

````python
results = collection.query(
    query_texts=["I love ipad"],
    n_results=2
)

print(f"\nQuery Results: {results}") # Look a the distance 'distances': [[1.619499683380127, 1.916476845741272]] they are further away. The far the distance is they are unrelated
````
