# Simple Tool Calling

````python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
````

````python
llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it", temperature=0) # temperature = 0 means no creativity, just factual answers. You can set it to 0.7 for more creativity.

response = llm.invoke("How many moons does Jupiter have?")
print(response.text)
````

````python
# Answer has to be in 1 single line is called System Instruction
response = llm.invoke([
    {"role": "system", "content": "You are a helpful assistant that answers questions about the solar system. You must answer in 1 single line."},
    {"role": "user", "content": "How many moons does Jupiter have?"}
])
print(response.text)
````

# Call LLM Using Groq Models

````python
from langchain_groq import ChatGroq
````

````python
llm = ChatGroq(model="qwen/qwen3-32b", reasoning_format="parsed")

messages = [
    {"role": "system", "content": "You are a helpful assistant that gives short answer to a question."},
    {"role": "user", "content": "How many moons does Jupiter have?"}
]

ai_message = llm.invoke(messages)
print(ai_message.content)
````
